MatplotLib is a plotting library for Python and its numerical mathematics extension NumPy. It provides an object oriented API for embedded plots into applications using general purpose GUI toolkits Tkinter, wxPython, Qt or GTK+.

Pros of MatplotLib:
    
    Generally easy to get started for simple plots 
    Support for custom labels and texts 
    Great control of every element in a figure
    High-quality output in many formats
    Very customizable in general

```python
import matplotlib.pyplot as plot

# The below line is used for Jupyter notebook, with this line in place we don't need to write plot.show() each time
%matplotlib inline
```

```python
import numpy as np
```

```python
# Simple example
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
# Plotting using matplotlib
plot.scatter(x,y,c='b') # c='b' represents the color of the graph
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.title("Graph in 2-D")
plot.savefig("First_Graph.png")
```

```python
y = x*x
```

```python
plot.plot(x, y,'ro--')
```

```python
# plot.subplot(rows, columns, graph_number) and then plot.plot()

plot.subplot(2,2,1)
plot.plot(x,y,'g--')

plot.subplot(2,2,2)
plot.plot(x,y,'r*-')

plot.subplot(2,2,3)
plot.plot(x,y,'bo')
```

```python
# Plotting a sine wave from 0 to 4pi at an interval of 0.1
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)

plot.title("Sine Wave Demo")
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.plot(x,y)
plot.show()
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plot.subplot(2,1,1)
plot.title("Sine Wave")
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.plot(x,y_sin)

plot.subplot(2,1,2)
plot.title("Cosine Wave")
plot.xlabel("X-Axis")
plot.ylabel("Y-Axis")
plot.plot(x,y_cos)

plot.show()
```

```python
# Bar graph demo
x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plot.bar(x1,y1,color="r", align="center")
plot.bar(x2,y2,color='g')
plot.title("Bar graph demo")
plot.xlabel("X Axis")
plot.ylabel("Y Axis")
```

```python
a = np.array([47, 56, 65, 74, 66, 98, 12, 7, 9, 45, 37, 87, 93, 75, 21])
plot.hist(a, bins = 20)  # default value for bins is 10
plot.title("Histogram Demo")
plot.xlabel("Values")
plot.ylabel("Density")
plot.plot()
```

# Box Plot Example

```python
data = [np.random.normal(0, standard_deviation, 100) for standard_deviation in range(1,4)]

# Rectangular box plot
plot.boxplot(data, vert=True, patch_artist=True)
plot.show()
```

# Pie Chart

```python
labels = ['Python', 'Java', 'Javascript', 'Ruby']
sizes = [200, 210, 180, 120]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0,0,0)      # explode represents how far away from the centre should a slice be placed

plot.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True) # autopct represents values to be shown in floating point format
plot.axis("Equal")
plot.show()
```
